```


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# database

data_path= '/content/database/Electric_Vehicle_Population_Data.csv'
ev_data=pd.read_csv(data_path)

#visualiza√ßao Dataframe
ev_data.head()
```

```
ev_data.info() #dataset info

ev_data.describe() #dataset description

#tipos de veiculos eletricos

sns.countplot(x='Electric Vehicle Type',data=ev_data)
plt.title('Electric Vehicle Type')
plt.show()
```

```
#tipo de veiculo eletrico por ano
plt.figure(figsize=(10, 6))
sns.countplot(x='Electric Vehicle Type', data=ev_data)
plt.title('Distribui√ß√£o dos Tipos de Ve√≠culos El√©tricos')
plt.xlabel('Tipo de Ve√≠culo El√©trico')
plt.ylabel('Contagem')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```
#Distribui√ßao por modelo de veiculo e por ano
plt.figure(figsize=(10, 6))
sns.histplot(ev_data['Model Year'], bins=20, kde=True)
plt.title('Distribui√ß√£o dos Ve√≠culos por Ano do Modelo')
plt.xlabel('Ano do Modelo')
plt.ylabel('Contagem')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
